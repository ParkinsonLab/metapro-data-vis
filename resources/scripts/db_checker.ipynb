```python
import sqlite3
import pandas as pd
import numpy as np
from pathlib import Path
from uuid import uuid4

db_path = Path('../db/taxonomy.db')
dmp_base_path = Path('../taxonomy')
# Notebook to check the database without risking the other ones

def query_df(query):
    with sqlite3.connect(db_path, autocommit=False) as connection:
        cursor = connection.cursor()
        cursor.execute(query)
        return cursor.fetchall()
```

```python
nodes_df = pd.read_csv(dmp_base_path / 'nodes.dmp', header=None, sep="|", usecols=[0, 1, 2], names = [
    'tax_id', 'parent_tax_id', 'rank'
], skipinitialspace=True, converters={'rank': lambda x: x.strip()})

names_df = pd.read_csv(dmp_base_path / 'names.dmp', header=None, sep="|", usecols=[0, 1, 3], names = [
    'tax_id', 'name', 'name_class'
], skipinitialspace=True, converters={'name': lambda x: x.strip(), 'name_class': lambda x: x.strip()})
```
